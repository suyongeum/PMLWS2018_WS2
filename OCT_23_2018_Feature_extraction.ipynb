# Feature extraction using panda

```
import numpy as np
import pandas as pd
```

```
url = 'https://raw.githubusercontent.com/suyongeum/PMLWS2018_WS2/master/data.csv'
traindata=pd.read_csv(url, encoding="shift-jis",header=None, names=('difficulty', 'word'),keep_default_na=False)
traindata.head()
```

# Feature adding

```
wordLength = []

for i in range(len(traindata)):
    wordLength.append(len(traindata.word[i]))
    
traindata["length"]=wordLength
traindata.head()
```

```
con_vol_ratio = []

for i in range(len(traindata)):
    w = traindata.word[i]
    vowel = w.count('a')+w.count('e')+w.count('i')+w.count('u')+w.count('o')
    consonant = len(w) - vowel + 1
    con_vol_ratio.append(vowel/consonant)
    
traindata["con_vol_ratio"]=con_vol_ratio
traindata.head()
```
